```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
from glob import glob
import glob

file_list = glob.glob("C:/Users/Nyamasi/OneDrive/Desktop/faam/*.csv") + glob.glob("C:/Users/Nyamasi/OneDrive/Desktop/faam/*.xlsx")+ glob.glob("C:/Users/Nyamasi/OneDrive/Desktop/faam/*.xls")
df_list =[]



#columns
columns= ['Price Period', 'Town','Super', 'Diesel','Kerosene']


#renaming  columns
column_mapping = {
    'Super ': 'Super',
    ' Diesel': 'Diesel',
    ' Diesel ': 'Diesel',
    'Kerosene ': 'Kerosene',
    'Petrol': 'Super',
    'Super Petrol': 'Super',
    'Period': 'Price Period',
    'TOWNS': 'Town',
    '                      TOWNS': 'Town'
}


for file in file_list:
    if file.endswith(".csv"):
        df = pd.read_csv(file)
    elif file.endswith(".xlsx"):
        df = pd.read_excel(file)
    elif file.endswith(".xls"):
        df = pd.read_excel(file)
        

        

    df_list.append(df)

# Concatenate all the DataFrames into a single DataFrame.
merged_df = pd.concat(df_list, ignore_index =True)



# Drop duplicate columns by keeping only the first occurrence
merged_df = merged_df.loc[:,~merged_df.columns.duplicated()]

# Select only the desired columns, if they exist in the DataFrame
merged_df = merged_df[[col for col in columns if col in merged_df.columns]]

# Save the cleaned and merged DataFrame to a new CSV file
merged_df.to_csv("merged_prices.csv", index=False)
```
